```python
import pymysql
import pandas as pd

#cursor = conn.cursor()

class MysqlSearch():
    def get_conn(self):
        #连接本地数据库
        try:
            self.conn = pymysql.connect(host = 'localhost'
                      ,user = 'root'
                      ,password = '123456'
                      ,database = 'im_ods_stage')
        except pymysql.error as e:
            print(e)
            print('数据库链接出错')
            
    def close_conn(self):
        #关闭数据库
        try:
            if self.conn:
                self.conn.close()
        except pymysql.Error as e:
            print(e)
            print('数据库关闭出错')
            
    def get_one(self):
        #查询一条数据
        try:
            #链接数据库
            self.get_conn()
            sql = 'select * from gdl_user_valid limit 10'
            cursor = self.conn.cursor()
            cursor.execute(sql)
            new = cursor.fetchone()
            # 返回一个字典，让用户可以按数据类型来获取数据
            new_dict = dict(zip([x[0] for x in cursor.description],new))
            cursor.close()
            self.close_conn()
            return new_dict
        except AttributeError as e:
            print(e)
            return None

    def get_all(self):
        '''获取所有结果'''
        sql = 'select * from gdl_user_valid limit 20 '
        self.get_conn()
        try:
            cursor = self.conn.cursor()
            cursor.execute(sql)
            news = cursor.fetchall()
            # 将数据转为字典，让用户根据键来查数据
            # news_list =list(map(lambda x:dict(zip([x[0] for x in cursor.description],[d for d in x])),news))
            # 这样也行,连续用两个列表生成式
            news_list = [dict(zip([x[0] for x in cursor.description],row)) for row in news]
            cursor.close()
            self.close_conn()
            return news_list
        except AttributeError as e:
            print(e)
            return None

    def get_more(self,page,page_size):
        '''查多少页的多少条数据'''
        offset = (page-1)*page_size
        sql = 'select * from gdl_user_valid LIMIT %s,%s'
        try:
            self.get_conn()
            cursor = self.conn.cursor()
            cursor.execute(sql,(offset,page_size,))
            news = [dict(zip([x[0] for x in cursor.description],new)) for new in cursor.fetchall()]
            cursor.close()
            self.close_conn()
            return news
        except AttributeError as e:
            print(e)
            return None

def main():
    news = MysqlSearch()
    new = news.get_more(6,4)
    if new:
        for row in new:
            print(row)
    else:
        print('获取数据失败')

if __name__ == '__main__':
    main()
```

```python
import pymysql

class OperateSQL():
    def get_conn(self):
        try:
            self.conn = pymysql.connect(database='im_ods_stage'
                                       ,host='localhost'
                                       ,user='root'
                                       ,password='123456')
        except pymysql.Error as e:
            print(e)
            print('数据库链接出错')
            
    def clo_conn(self):
        try:
            if self.conn:
                self.conn.close() 
        except pymysql.Error as e:
            print(e)
            print('数据库关闭出错')
            
    def add_one(self):
        sql = 'insert into abab(name,address) values(%s,%s)'
        try:
            self.get_conn()
            cursor = self.conn.cursor()
            #print(sql,('python测试','第二个'))
            cursor.execute(sql, ('python测试', '第二个'))
            
        # 一定需要提交事务，要不不会显示，只会占位在数据库
            self.conn.commit()
            print('数据写入成功')
            return 1
        except AttributeError as e:
            print('Error:', e)
            return 0
        except TypeError as e:
            print('Error:', e)
            # 发生错误还提交就是把执行正确的语句提交上去
            # self.conn.commit()
            # 下面这个方法是发生异常就全部不能提交,但语句执行成功的就会占位
            self.conn.rollback()
            return 0
        finally:
            cursor.close()
            self.clo_conn()
            
def main():
    news = OperateSQL()
    data = news.add_one()
    if data:
        print('数据写入成功')

if __name__ == '__main__':
    main()
```

```python
import pymysql

class OperateSQL():
    def get_conn(self):
        try:
            self.conn = pymysql.connect(database='im_ods_stage'
                                       ,host ='localhost'
                                       ,user = 'root'
                                       ,password = '123456')
        except pymysql.Error as e:
            print(e)
            print('数据库链接异常')
            
    def clo_conn(self):
        try:
            if self.conn:
                self.conn.close()
                
        except pymysql.Error as e:
            print(e)
            print('数据库链接异常')
    
    def add_one(self):
        try:
            self.get_conn()
            cursor = self.conn.cursor()
            sql = 'insert into abab(name,address) values(%s,%s)'
            cursor.execute(sql,('第二次python链接','测试数据'))
            self.conn.commit()
        except AttributeError as e:
            print(e)
        except TypeError as e:
            print(e)
        finally:
            cursor.close()
            self.clo_conn()
            print('数据写入成功！')
    
    def del_one(self):
        try:
            self.get_conn()
            cursor = self.conn.cursor()
            sql = 'delete from abab where address = %s'
            cursor.execute(sql,('测试数据',))
            self.conn.commit()
            print('数据删除成功')
        except pymysql.Error as e:
            print(e)
            print('数据删除失败')
            if self.conn:
                self.conn.rollback()
        finally:
            if self.conn:
                self.clo_conn()
                
    def upd_one(self):
        try:
            self.get_conn()
            cursor = self.conn.cursor()
            sql = 'update abab set name="修改后数据" where address= "测试数据" '
            cursor.execute(sql)
            self.conn.commit()
            print('数据修改成功')
        except pymysql.Error as e:
            print(e)
        finally:
            cursor.close()
            self.clo_conn()
    
def main():
    news = OperateSQL()
    news.upd_one()
    
if __name__ == '__main__':
    main()
```
